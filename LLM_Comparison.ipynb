```python
# =========================
# Compare All LLMs: BioGPT vs LLaMA vs Gemma vs Meditron
# Metrics are for records 21-35
# Positive class: Radiotherapy
# =========================

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Extracted metrics
df_metrics = pd.DataFrame({
    "Model": ["BioGPT", "LLaMA-3.1", "Gemma", "Meditron"],
    "Accuracy":  [0.8667, 0.9333, 0.4000, 0.6000],
    "Precision": [1.0000, 1.0000, 0.4000, 0.0000],
    "Recall":    [0.6667, 0.8333, 1.0000, 0.0000],
    "F1-score":  [0.8000, 0.9091, 0.5714, 0.0000],
})

print(df_metrics)

# -------------------------
# Grouped bar chart
# -------------------------
metrics = ["Accuracy", "Precision", "Recall", "F1-score"]
x = np.arange(len(metrics))
width = 0.2

plt.figure(figsize=(11, 6))

for i, row in df_metrics.iterrows():
    plt.bar(
        x + (i - 1.5) * width,
        [row[m] for m in metrics],
        width=width,
        label=row["Model"]
    )

plt.xticks(x, metrics)
plt.ylim(0, 1.1)
plt.ylabel("Score")
plt.title("Comparison of LLM Performance on Radiotherapy Prediction (Records 21–35)")
plt.legend()
plt.grid(axis="y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

models = ["BioGPT", "LLaMA-3.1", "Gemma", "Meditron"]

accuracy  = [0.8667, 0.9333, 0.4000, 0.6000]
precision = [1.0000, 1.0000, 0.4000, 0.0000]
recall    = [0.6667, 0.8333, 1.0000, 0.0000]
f1        = [0.8000, 0.9091, 0.5714, 0.0000]

data = np.array([accuracy, precision, recall, f1]).T
labels = ["Accuracy", "Precision", "Recall", "F1-score"]

angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
angles = np.concatenate((angles, [angles[0]]))

plt.figure(figsize=(6,6))
for i, model in enumerate(models):
    values = np.concatenate((data[i], [data[i][0]]))
    plt.polar(angles, values, marker='o', label=model)

plt.xticks(angles[:-1], labels)
plt.title("LLM Performance Radar Chart")
plt.legend(loc="upper right")
plt.show()
```
